```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
mpl.rcParams["patch.force_edgecolor"]=True
mpl.rcParams["patch.edgecolor"]="Green"
mpl.rcParams["patch.linewidth"]=3
```

```python
sns.distplot(tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.distplot(tips["tip"])
#sns.distplot(tips["tip"],kde=False)#if u dont want the line distribution
plt.title("Distribution of tips")
plt.xlabel("Tips")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='reg')#kind=(default)scatterplot,reg,hex,kde
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
sns.kdeplot(tips['tip'])#kernel density estimation plot
plt.show()
```

```python
grid= sns.FacetGrid(tips,col='time',row='smoker')
grid.map(plt.hist,'total_bill')
plt.show()
```

```python
grid= sns.FacetGrid(tips,col='time',row='smoker')
grid.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
grid=sns.JointGrid(x="total_bill",y="tip",data=tips)
grid.plot(plt.scatter,sns.distplot)
plt.show()
```
