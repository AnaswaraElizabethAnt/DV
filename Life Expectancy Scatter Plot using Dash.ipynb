```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
life = pd.read_csv('D:/dataset/Life Expectancy Data.csv')
```

```python
life.head()
```

```python
life.Year.unique()
```

```python
life.describe()
```

```python
app = dash.Dash()

app.layout = html.Div([ # this Div is the main Div containing all charts and html components
    html.H1('Life Expectancy Dashboard',
           style={'textAlign': 'center',
                  'color': 'green'
        }),
    html.Div([html.H5("Level"),
        dcc.Slider(id='slider_input',
            min=2000,
            max=2015,
            step=1,
            value=2000,
            marks={i: '{}'.format(i) for i in [2000,2001,2002,2003,2004,2005,2006,2007,
                                                    2008,2009,2010,2011,2012,2013,2014,2015]},
        ),
    ],style={'height': '100','width':'90%','paddingLeft':'30px'}),
    dcc.Graph(
        id='my_scatter',
        animate=True
    )
])

@app.callback(Output('my_scatter', 'figure'),
             [Input('slider_input', 'value')])

def update_plot(slider_input):
    life_year = life[life['Year']==slider_input]
    trace1 = go.Scatter(
        x = life_year[life_year['Status']=='Developing']['Life expectancy '],
        y = life_year[life_year['Status']=='Developing']['GDP'],
        mode = 'markers',
        name = 'Developing',
        marker = dict(
            size = 13,
            color = 'lightgreen'
        )
    )
    trace2 = go.Scatter(
        x = life_year[life_year['Status']=='Developed']['Life expectancy '],
        y = life_year[life_year['Status']=='Developed']['GDP'],
        mode = 'markers',
        name = 'Developed',
        marker = dict(
            size = 13,
            color = 'pink'
        )
    )
    data = [trace1,trace2]
    
    return {
        'data': data,
        'layout': go.Layout(
            autosize=False,
            title='Scatter plot',
            hovermode='closest',
            yaxis=dict(zeroline=False, range=[-10000,50000]),
        )
            
    }
```

```python
if __name__ == '__main__':
    app.run_server()
```
