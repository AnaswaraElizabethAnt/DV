```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mov=pd.read_csv("D:/dataset/tmdb_5000_movies.csv")
```

### draw a correlogram of movies dataset ignoring id column

```python
mov1=mov.drop(labels=None, axis=0, index=None, columns='id', level=None, inplace=False, errors='raise')
```

```python
mov1.shape
```

```python
sns.heatmap(mov1.corr(),annot=True,cmap="coolwarm")
plt.title("correlation of the movie dataset")
plt.show()
```

### draw a cluster map of the corelation of all numeric column except the id column.what do u observe?

```python
grid=sns.PairGrid(mov1)
grid.map(plt.scatter)
plt.show()
```

```python
mpl.rcParams["patch.force_edgecolor"]=True
mpl.rcParams["patch.edgecolor"]="Green"
mpl.rcParams["patch.linewidth"]=3
```

```python
grid=sns.PairGrid(mov1)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.regplot)#kdeplot
plt.show()
```

```python
sns.clustermap((mov1.corr()),annot=True,cmap='rainbow')
plt.show()
```

###show the top 10 movies based on popularity whose vote_average is between a)4 to 6 b)6 to 8 c)8 to 10 

```python
mov.head()
```

```python
mov4to6=mov.query("vote_average>4 and vote_average<6").groupby(['popularity','original_title']).size()
mov4to6=pd.DataFrame(mov4to6)
mov4to6=mov4to6.reset_index()
mov4to6=mov4to6.tail(10)
mov4to6
```

```python
fig=plt.figure(figsize=(12,7))
sns.barplot(x='original_title',y='popularity',data=mov4to6)
plt.ylabel('popularity')
plt.title("top 10 movies based on popularity where vote avg is between 4 to 6")
plt.xlabel("movies")
plt.xticks(rotation='vertical')
plt.show()
```
