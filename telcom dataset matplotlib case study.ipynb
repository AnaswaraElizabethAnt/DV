```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tel=pd.read_csv("D:/dataset/Telco-Customer-Churn.csv")
```

```python
tel100=tel.head(100)
tel100
```

```python
tel.shape
tel.describe()
```

```python
#how is the distributon of tenure?

plt.hist(tel['tenure'],color='orange',edgecolor='black')
plt.xlabel("tenure")
plt.ylabel("count of tenure")
plt.title("distribution of tenure")
plt.show()
```

```python
#how is the distribution of monthly charges?

plt.hist(tel['MonthlyCharges'],color='orange',edgecolor='black')
plt.xlabel("Monthly Charges")
plt.ylabel("count of Monthly Charges")
plt.title("distribution of Monthly Charges")
plt.show()
```

```python
pym=tel.groupby(['InternetService']).size()
pym=pd.DataFrame(pym)
pym=pym.reset_index()
pym=pym.rename(columns={0:'Count'})
pym
```

```python
#what is the part to whole relationship of internet services?

plt.pie(pym['Count'],labels=pym['InternetService'],autopct='%1.2f%%',startangle=90)
plt.axis('equal')
plt.show()
```

### which contract is most common?what is the percentage of most common contract?

```python
con=tel.groupby(['Contract']).size()
con=pd.DataFrame(con)
con=con.reset_index()
con=con.rename(columns={0:'Count'})
con
```

```python
plt.pie(con['Count'],labels=con['Contract'],autopct='%1.2f%%',startangle=90,explode=(0.1,0.2,0),pctdistance=0.5)
plt.axis("equal")
plt.show()
```

```python
pmt_method=tel.groupby(['PaymentMethod']).size()
pmt_method=pd.DataFrame(pmt_method)
pmt_method=pmt_method.reset_index()
pmt_method=pmt_method.rename(columns={0:'Count'})
```

```python
col = ['red','green','yellow','blue']
plt.pie(pmt_method['Count'],labels=pmt_method['PaymentMethod'],labeldistance=0.9,
       autopct='%0.1f%%',startangle=90,wedgeprops=dict(width=0.5),radius=1)
plt.axis("equal")
plt.show()
```

```python
pmt_by_gdr=tel.groupby(['PaymentMethod','gender']).size()
pmt_by_gdr=pd.DataFrame(pmt_by_gdr)
pmt_by_gdr=pmt_by_gdr.reset_index()
pmt_by_gdr=pmt_by_gdr.rename(columns={0:'Count'})
```

```python
l=['coral','crimson','goldenrod','orchid']
plt.pie(pmt_by_gdr[pmt_by_gdr['gender']=='Male']['Count'],
       autopct='%0.1f%%',pctdistance=0.8,startangle=90,
        wedgeprops=dict(width=0.5),radius=1,colors=col)
plt.pie(pmt_by_gdr[pmt_by_gdr['gender']=='Female']['Count'],
       autopct='%0.1f%%',pctdistance=1.1,startangle=90,
        wedgeprops=dict(width=0.5),radius=1.55,colors=col)
plt.title("variation by payment method",loc='center',color='maroon')
plt.legend(labels=pmt_by_gdr[pmt_by_gdr['gender']=='Female']['PaymentMethod'],
          loc="right",bbox_to_anchor=(1.4,0.7))
plt.axis('equal')
plt.show()
```
